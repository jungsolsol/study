##객체 지향 프로그래밍

#객체들의 모임으로 프로그램을 구성 각각 개체들이 메세지를 주고받고 데이터 처리
#객체지향 프로그래밍은 유연하고 변경이 용이함(키보드 마우스 교체하듯 컴포넌트를 쉽게 변경가능)

#다형성#
-역할과 구현으로 구분
장-> 클라이언트는 대상의 역할만 알면됨 내부구조몰라도되고 변경되어도 상관없고 대상자체를 변경해도 괜찮음
->자바언어 역할 = 인터페이스, 구현= 인터페이스를 구현한 클래스, 구현객체
오버라이딩 - 메소드 오버라이딩이란 상속받은 부모 클래스의 메소드를 재정의하여 사용하는 거
한계 -> 인터페이스 자체가 변하면 큰변경발생(인터페이스를 잘 설계하는게 중요하다)

#스프링과 객체지향
다형성이 가장 중요함
스프링은 다형성을 극대화해서 이용할수있게 도와준다.(DI~~)

#좋은 객체지향설계 5원칙 SOLID

SSP 단일책임원칙
-한 클래스는 하나의 책임만 가져야함(좀모호하긴하다) -변경이 있을때 파급효과가 적으면 SSP를 잘따른것

OCP *개방 폐쇠원칙
소프트웨어 요소는 확장에는 열려있으나 변경에는 닫혀있어야한다. 다형성!
문제점 구현객체를 변경하려면 클라이언트 코드를 변경해야함 Member m = new MemberService();(변경전)
Member m = new JdbcService();(변경후 Member 코드 변경됨)-> 별도의 조립, 설정자가 필요(SPRING)

LSP 리스코프 치환원칙
인터페이스의 구현체는 기능적 규약을 맞춰야함(엑셀인데 뒤로가면 LSP위반)

ISP 인터페이스 분리원칙
자동차 인터페이스 -> 운전 인터페이스, 정비 인터페이스로 분리 
==> 사용자 클라이언트 => 운전자 클라이언트, 정비사 클라이언트로 분리가능

DIP *의존관계 역전 원칙
구현체에 의존하지말고 인터페이스에 의존해야함 -> 역할에 의존해야함

*다형성만으로는 OCP, DIP를 지킬수없음 =>스프링을써야함

이상적으론 모든설계에 인터페이스를 부여하자

##비즈니스 요구사항과 설계
#회원
    회원을 가입하고 조회할 수 있다.
    회원은 일반과 VIP 두 가지 등급이 있다.
    회원 데이터는 자체 DB를 구축할 수 있고, 외부 시스템과 연동할 수 있다. (미확정)
#주문과 할인 정책
    회원은 상품을 주문할 수 있다.
    회원 등급에 따라 할인 정책을 적용할 수 있다.
    할인 정책은 모든 VIP는 1000원을 할인해주는 고정 금액 할인을 적용해달라. (나중에 변경 될 수
    있다.)
    할인 정책은 변경 가능성이 높다. 회사의 기본 할인 정책을 아직 정하지 못했고, 오픈 직전까지 고민을
    미루고 싶다. 최악의 경우 할인을 적용하지 않을 수 도 있다. (미확정)
